```python
import pandas as pd
import numpy as np
import pyproj
from pyproj import Transformer
```

```python
main = pd.read_excel('/content/drive/MyDrive/공유폴더/NEW(공모전용)/데이터위도경도포함.xlsx',index_col=0)
main.head(1)
```

```python
# 서울역 위도 경도 (37.553230, 126.972551) (예시)
transformer = Transformer.from_crs("epsg:4326", "epsg:5178")
transformer.transform(37.553230, 126.972551)
```

```python
# transformer.transform(37.560927, 127.193877)
list1 = []
for i in range(0, len(main)):
    list1.append(transformer.transform(main['위도'][i], main['경도'][i]))

list1
```

```python
df = pd.DataFrame(list1)
df.columns = ['위도UTMK', '경도UTMK']
df
```

```python
main0 = pd.concat([main, df], axis=1)
main0
```

```python
main0.to_excel('주민센터 UTMK변환.xlsx')
```

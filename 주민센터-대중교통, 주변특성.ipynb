# 주변특성 시작

```python
library = pd.read_excel('/content/drive/MyDrive/공유폴더/NEW(공모전용)/대중교통, 주변특성 데이터/공공도서관 위도경도 변환.xlsx',index_col =0)
library = library.iloc[:,13:15]
library.head(1)
```

```python
x = main['위도UTMK']
x.head(1)
```

```python
y = main['경도UTMK']
y.head(1)
```

```python
a1 = library['위도UTMK']
a1.head(1)
```

```python
a2 = library['경도UTMK']
a2.head(1)
```

```python
main["도서관(500)"] = 0
main.head(1)
```

```python
for i in range(len(main)):
    for j in range(len(library)):
        value = False
        value = np.sqrt((x[i] - a1[j])**2 + (y[i] - a2[j])**2) <= 500
        if value == True:
            main['도서관(500)'][i] += int(value)
        else:
            pass
main
```
